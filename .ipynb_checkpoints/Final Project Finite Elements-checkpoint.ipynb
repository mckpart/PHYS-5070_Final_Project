```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import analytic_sol 
```

# $\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = \nu \frac{\partial^2 u}{\partial x^2}$ 

```python
def FiniteElements_Burgers(v,grid,time,initial_conditions):
    delta_x = grid[1] - grid[0]
    delta_t = time[1] - time[0]
    
    u = [initial_conditions]
    
    dudt = np.empty(len(grid))
    for t in time[1:]:
        dudt[0] = 0
        dudt[-1] = 0
        for n in range(1,len(grid)-1):
            dudt[n] = v*(u[-1][n+1] - 2*u[-1][n] + u[-1][n-1])/delta_x**2 - u[-1][n]*(u[-1][n+1] - u[-1][n-1])/(2*delta_x)
            #dudt[n] = v*(u[-1][n+1] - 2*u[-1][n] + u[-1][n-1])/delta_x**2
            
        u.append(u[-1] + dudt*delta_t)
        
    return np.array(u)
```

```python
grid = np.linspace(0,1,20)
initial = np.sin(np.pi*grid)
time=np.linspace(0,1,10000)

u = FiniteElements_Burgers(v=1,grid=grid,time=time,initial_conditions=initial)

plt.plot(grid,u[0])
```

```python
Grid,Time = np.meshgrid(grid,time)

print(u)

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Grid,Time,u)
plt.show()
```
